```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv("templates/timeslots.csv",parse_dates=['Day', 'Timestamp UTC ms', 'Time'])
df.head()
```

```python
start = datetime(2018,7,29)
end = datetime(2018,7,31)
```

```python
df_ = df[(df['Day']>start) & (df['Day']<end)]
df_.drop(columns=['Timestamp UTC ms', 'Steps', 'Move'])
df_['Duration_min'] = df_['Duration ms']/(60000*60)
#df_ = df_[df_['Activity']!=nan]
df_.head()
```

```python
pd.isnull(df_['Activity'][869])
```

```python
def act_split(strings, index=0):
    if pd.isnull(strings):
        return 'NaN'
    strings = strings.split(': ')
    if len(strings)==1:
            return strings[0]
    return strings[index]
```

```python
df_['act1']= df_['Activity'].apply(act_split)
df_['act2']= df_['Activity'].apply(act_split, index=1)
```

```python
def activity_data(df, n=20):
    _ = []
    for i in range(n):
        children = dict({'name': df.iloc[i].act2,
        'value': df.iloc[i].Duration_min})
        _.append(dict({'name': df.iloc[i].act1,
        'children': [children]}))
    return _
```

```python
import json
json.dumps(activity_data(df_,len(df_)))
with open('templates/data.json', 'w') as f:
    json.dump(activity_data(df_,len(df_)), f)
```
